# TOP 50 IMDB RATING MOVIE ON IMDB WEBSAITE

```python
import requests
from bs4 import BeautifulSoup
url="https://www.imdb.com/search/title/?release_date=2017-01-01,2017-12-31&sort=num_votes,desc&start=&ref_=adv_prv"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
```

```python
head=soup.find(class_ ="header")
head.get_text()
```

```python
data=soup.find_all(class_="lister-item mode-advanced")
data
len(data)
```

```python
rating=soup.find(class_= "inline-block")
rating.get_text().replace('\n','').split()
```

```python
Catagory=soup.find('p')
Catagory.get_text().replace('\n','').split()
```

```python
votegross=soup.find(class_ ="sort-num_votes-visible")
votegross.get_text().replace('\n','').split()
```

```python
title,Rating,Catogary,votegross=[],[],[],[]
for i in data:
        title.append(i.find(class_ ="lister-item-header").get_text().replace('\n',''))
        Rating.append(i.find(class_= "inline-block ratings-imdb-rating").get_text().replace('\n','').split())
        Catogary.append(i.find('p').get_text().replace('\n',''))
        votegross.append(i.find(class_ ="sort-num_votes-visible").get_text().replace('\n','').split())
```

```python
len(title),len(Rating),len(Catogary),len(votegross)
```

```python
import pandas as pd
d={'title':title,"Rating":Rating,'Catogary':Catogary,'Votegross':votegross}
df=pd.DataFrame(d)
df.sort_values('Rating',ascending=False)
```

```python
df.to_csv('TOP_50_IMDB.csv')
```
