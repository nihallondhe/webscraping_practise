```python
import requests
```

```python
from bs4 import BeautifulSoup
url=f"https://www.metoffice.gov.uk/pub/data/weather/uk/climate/datasets/Raindays1mm/ranked/UK.txt"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
soup=list(soup)
sou=str(soup)
```

```python
import re 
rs=re.findall(r'(\d+.\d+\s+\d{4})',sou)
ps=re.findall(r'(\w{3}\s+year)',sou)
ps
```

```python
#            dit.clear()
dit={}
for i in ps:
    dit[i]=[] #Append Keys in Dict
for i in range(1,len(rs)-2,17):  # For exact Keys Otherwise we can iterate on Dict of keys:
    dit['jan  year'].append(rs[i])
    dit['feb  year'].append(rs[i+1])
    dit['mar  year'].append(rs[i+2])
    dit['apr  year'].append(rs[i+3])
    dit['may  year'].append(rs[i+4])
    dit['jun  year'].append(rs[i+5])
    dit['jul  year'].append(rs[i+6])
    dit['aug  year'].append(rs[i+7])
    dit['sep  year'].append(rs[i+8])
    dit['oct  year'].append(rs[i+9])
    dit['nov  year'].append(rs[i+10])
    dit['dec  year'].append(rs[i+11])
    dit['win  year'].append(rs[i+12])
    dit['spr  year'].append(rs[i+13])
    dit['sum  year'].append(rs[i+14])
    dit['aut  year'].append(rs[i+15])
    dit['ann  year'].append(rs[i+16])
```

```python
dit
```

```python
import pandas as pd
df=pd.DataFrame(dit)
```
